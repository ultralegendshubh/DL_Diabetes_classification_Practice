```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
from tensorflow.keras.utils import plot_model
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/drive/MyDrive/diabetes.csv")
df
```

```python
df.shape
```

```python
x=df.drop(columns=["Outcome"],axis=1)
y=df["Outcome"]
```

```python
x.shape,y.shape
```

```python
# spliting the data set in train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
x_train.shape,x_test.shape,y_test.shape,y_train.shape
```

```python
 # scaling the data set to avoid
```

```python
scaler=StandardScaler()
x_train_scaled=pd.DataFrame(scaler.fit_transform(x_train),
                            columns=x_train.columns)
x_test_scaled=scaler.transform(x_test)
```

```python
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_train.columns)
```

```python
x_train_scaled,x_test_scaled
```

```python
display(x_test_scaled)
```

#Building the model Architecture

```python
# creating the instaance of model
model=tf.keras.Sequential(name="Main_Container")

# Adding the input layer to model
model.add(tf.keras.layers.Input(shape=[8],name="Input_layer"))
# creating the 1st hidden layer
model.add(tf.keras.layers.Dense(units=9,activation="relu",name="hidden_layer_1"))

# creating the 2nd hidden layer
model.add(tf.keras.layers.Dense(units=8,activation="relu",name="hidden_layer_2"))
# creating the 3rd hidden layer
model.add(tf.keras.layers.Dense(units=4,activation="relu",name="hidden_layer_3"))
# create the output layer
model.add(tf.keras.layers.Dense(units=1,activation="sigmoid",name="output_layer"))
```


# weight initiliaze


```python
model.layers[0].get_weights()
```

```python
model.summary()
```

```python
plot_model(model,show_shapes=True,show_layer_names=True,show_layer_activations=True)
```

```python
# once the architecheture is being created next step is to add functionlity to the model
model.compile(
    loss="binary_crossentropy",
              optimizer="adam",
              metrics=['accuracy'])
```

```python
# now lets fit the data
history=model.fit(x_train_scaled,y_train,
                  epochs=100,batch_size=16,validation_split=0.2,workers=10,
                  use_multiprocessing= True,verbose=True)
```
